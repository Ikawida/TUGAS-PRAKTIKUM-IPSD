2. Distribusi Kategori: menghitung frekuensi tiap kategori (seperti pendidikan, status pernikahan) di kedua dataset dan membandingkan distribusinya.

```python
# 2. Distribusi Kategori

# Kolom kualitatif yang ingin dibandingkan
categorical_cols = ['pendidikan', 'status_pernikahan']

# Distribusi untuk data test_features 2
test_categorical = {
    col: data_test_features_2[col].value_counts(dropna=False) for col in categorical_cols
}

# Distribusi untuk data train_features 1
train_categorical = {
    col: data_train_features_1[col].value_counts(dropna=False) for col in categorical_cols
}

print("\nDistribusi Kategori Data Test_Features 2:")
for col, dist in test_categorical.items():
    print(f"\n{col}:\n{dist}")

print("\nDistribusi Kategori Data Train_Features 1:")
for col, dist in train_categorical.items():
    print(f"\n{col}:\n{dist}")
```

3. Visualisasi dan Uji T-Test dengan menggunakan data pendapatan dari kedua dataset (data test berwarna biru dan data train berwarna merah) dalam satu grafik.

```python
# 3. Visualisasi Data Pendapatan

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Membuat histogram untuk kolom 'pendapatan' dari data test_features 2
# Menggunakan warna biru dan menambahkan garis KDE (Kernel Density Estimate)
sns.histplot(data_test_features_2['pendapatan'], color='blue', label='Data Test', kde=True)

# Membuat histogram untuk kolom 'pendapatan' dari data train_features 1
# Menggunakan warna merah dan menambahkan garis KDE
sns.histplot(data_train_features_1['pendapatan'], color='red', label='Data Train', kde=True)

plt.title('Distribusi Pendapatan') # judul untuk grafik
plt.xlabel('Pendapatan') # label untuk sumbu x
plt.ylabel('Frekuensi') # label untuk sumbu y
plt.legend() # Menampilkan legenda untuk membedakan antara data test dan data train
plt.show()  # Menampilkan grafik
```

```python
# 4. Uji T-Test Data Pendapatan

from scipy import stats

# Ambil kolom 'pendapatan' dari kedua dataset
pendapatan_train = data_train_features_1['pendapatan']
pendapatan_test = data_test_features_2['pendapatan']

# Lakukan independent t-test
t_stat, p_value = stats.ttest_ind(pendapatan_train, pendapatan_test)

# Cetak hasil uji
print(f"T-Statistik: {t_stat}")
print(f"P-Value: {p_value}")

# Tentukan apakah ada perbedaan yang signifikan
alpha = 0.05  # Level signifikansi 5%

if p_value < alpha:
    print("Terdapat perbedaan yang signifikan antara distribusi pendapatan di data train dan data test.")
else:
    print("Tidak terdapat perbedaan yang signifikan antara distribusi pendapatan di data train dan data test.")
```

INTERPRETASI:

Dari hasil visulaisasi data terlihat bahwa karakteristik pendapatan di kedua dataset test dan train tidak jauh berbeda, baik dalam distribusi, rentang, maupun kepadatan. Kemudian dari hasil uji T-Test variabel pendapatan pun terlihat bahwa
tidak terdapat perbedaan yang signifikan antara distribusi pendapatan di data train_features 1 dan data test_features 2.

Jadi, dapat disimpulkan bahwa dari hasil visualisasi dan uji T-Test variabel pendapatan kedua dataset tidak terdapat perbedaan yang signifikan.
